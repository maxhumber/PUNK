### Cryptopunk Transaction EDA

```python
import pandas as pd
import numpy as np
import altair as alt
```

All transactions until May 16, 2021:

```python
tdf = pd.read_csv("data/transactions.csv")
tdf["date"] = pd.to_datetime(tdf["date"])
```

```python
tdf.shape
```

```python
tdf.head()
```

Zoom in on one:

```python
punk = 6601
pdf = tdf[tdf["id"] == punk]
```

```python
pdf
```

```python
pdf = pdf.dropna(subset=["eth"])
```

```python
alt.Chart(pdf).encode(x="date", y="eth", color="type").mark_circle()
```

```python
pdf = pdf[pdf["date"] >= "2021-01-01"]
```

```python
(
    alt.Chart(pdf)
    .encode(x="date", y="eth", color="type")
    .mark_circle()
    .properties(title='Punk 6601')
)
```

Quick look general trend:

```python
sold = tdf[tdf["type"] == "Sold"].sample(5000)
```

```python
alt.Chart(sold).encode(x="date", y="usd").mark_circle()
```

```python
sold["eth_log"] = np.log(sold["eth"])
```

```python
alt.Chart(sold).encode(x="date", y="eth_log").mark_circle()
```

Crude analysis based on attributes:

```python
adf = pd.read_csv("data/attributes.csv")
```

```python
sold90 = sold[(pd.Timestamp("now") - sold["date"]).dt.days <= 90]
sold90 = sold90[["id", "eth", "date"]]
df = pd.merge(sold90, adf, on="id")
df = df.dropna()
```

```python
df
```

```python
df['attributes'] = df['attributes'].apply(lambda x: x.split(' + '))
ethattr = df.explode('attributes')
ethattr
```

```python
ethattr = ethattr.groupby("attributes")["eth"].median().reset_index()
```

```python
alt.Chart(ethattr).encode(x="eth", y="attributes").mark_bar()
```
