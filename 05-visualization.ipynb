```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
hdf = pd.read_csv("csv/history.csv")
hdf["date"] = pd.to_datetime(hdf["date"])
```

```python
hdf.head()
```

```python
sold = hdf[hdf["type"] == "Sold"].sample(5000)
```

```python
alt.Chart(sold).encode(x="date", y="usd").mark_circle()
```

```python
sold["usd_log"] = np.log(sold["usd"])
```

```python
alt.Chart(sold).encode(x="date", y="usd_log").mark_circle()
```

```python
sold["eth_log"] = np.log(sold["eth"])
```

```python
alt.Chart(sold).encode(x="date", y="eth_log").mark_circle(color="green")
```

```python
sold["ym"] = sold['date'].apply(lambda x: f"{x.year + x.month/12}")
```

```python
sold = sold[sold["usd"] > 0]
```

```python
sold_ym = sold.groupby("ym")[["eth_log", "usd_log"]].median().reset_index()
sold_ym["ym"] = sold_ym["ym"].apply(float)
```

```python
sold_ym
```

```python
alt.Chart(sold_ym).encode(x="ym", y="usd_log").mark_line()
```
